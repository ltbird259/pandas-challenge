```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
#renaming data and ensuring it is df
data = pd.DataFrame(purchase_data)
data
#setting the data to be what we need it to be to count unique players
uniqueSN = data['SN'].unique()
uniqueSNseries = pd.Series(uniqueSN)
countuniquesn = uniqueSNseries.count()
countuniquesn
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#average price of items
avgprice = data['Price'].mean()
avgprice
#unique items sold
uniqueitems = data['Item Name'].nunique()
uniqueitems
#total purchases made
totalpurchases = data['Purchase ID'].count()
totalpurchases
#total revenue
totalprice = data['Price'].sum()
totalprice
dfsummary = pd.DataFrame([{ 'Average Price' : avgprice,
                          'Unique Items' : uniqueitems,
                          'Total Purchases': totalpurchases,
                          'Total Price' : totalprice
    }])
dfsummary
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#count unique males
males = data.loc[data["Gender"] == "Male", :]
uniquemales = males["SN"].nunique()
uniquemales
#count unique females
females = data.loc[data["Gender"] == "Female", :]
uniquefemales = females["SN"].nunique()
uniquefemales
#count of other
other = data.loc[data["Gender"] == "Other / Non-Disclosed", :]
uniqueother = other["SN"].nunique()
uniqueother
#creating df
genderdemo = pd.DataFrame({'Gender': ['Male', 'Female', 'Other'],
                           'Total Count': [uniquemales,uniquefemales,uniqueother]
                            
                           })
genderdemo
#adding the percent column
genderdemo['Percent of players'] = [str(round(x*100/countuniquesn,2))+"%" for x in genderdemo['Total Count']]
genderdemo.set_index('Gender', inplace=True)
genderdemo
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
data2 = data[['SN','Gender']]
```

```python
#gender group established
gendergroup = data.groupby(['Gender'])
#gender purchasecount
gendergroupcount = gendergroup['Purchase ID'].count()
gendergroupcount
#gender average
gendergroupavg = gendergroup['Price'].mean()
gendergroupavg
#gender totals
gendergrouptotal = gendergroup['Price'].sum()
gendergrouptotal
#gender purchase total avg per person
#first group by person and sum
avgper = data.groupby(['SN']).sum() 
avgper
#merge gender back to new df
genderavgper = avgper.merge(data2[['SN','Gender']], on='SN')
#remove duplicates
genderavgper2 = genderavgper.drop_duplicates()
genderavgper2
#group by gender for avg price total per person
genderavgperfinal = genderavgper2.groupby('Gender')['Price'].mean()
genderavgperfinal
# create final df
genderanalysis = pd.DataFrame({'Purchase Count': gendergroupcount,
                               'Average Purchase Price': gendergroupavg,
                               'Total purchase Value': gendergrouptotal,
                               'Avg Total Purchase per Person': genderavgperfinal   
                                })
genderanalysis
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# was reading wrong instructions. Check below for the age demo information
#creating the bins
bins = [-1,9,14,19,24,29,34,39,150]
binnames = ["<10",'10-14','15-19','20-24','25-29','30-34','35-39','40+']
data['Age bins'] = pd.cut(data['Age'], bins, labels=binnames)
#group by
agegroup = data.groupby('Age bins')
# count of age
agegroupcount = agegroup['Purchase ID'].count()
agegroupcount
# percent
agegrouppercent = [round(y*100/countuniquesn,2) for y in agegroupcount]
agegrouppercent
# sum
agegrouptotal = agegroup['Price'].sum()
agegrouptotal
# pulled in avg per from prev block.
ageavgper = avgper.merge(data[['SN','Age bins']], on='SN')
ageavgper
ageavgper2 = ageavgper.drop_duplicates()
ageavgperfinal = ageavgper2.groupby('Age bins')['Price'].mean()
ageavgperfinal
#creating final df
ageanalysis = pd.DataFrame({'Purchase Count': agegroupcount,
                               'Average Purchase Price': agegrouppercent,
                               'Total purchase Value': agegrouptotal,
                               'Avg Total Purchase per Person': ageavgperfinal   
                                })
ageanalysis
```

```python
uniqueplayers = data.drop_duplicates('SN')
uniqueplayers
playerage = uniqueplayers.groupby('Age bins')
playeragecount = playerage['Purchase ID'].count()
playeragepercent = [round(z*100/countuniquesn,2) for z in playeragecount]
playeragesummary = pd.DataFrame({'Total Count': playeragecount,
                                 'Percentage of Players': playeragepercent})
playeragesummary
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting

```python
playergroups = data.groupby('SN')
playergroupscount = playergroups['Purchase ID'].count()
playergroupsavg = playergroups['Price'].mean()
playergroupssum = playergroups['Price'].sum()
playergroupssummary = pd.DataFrame({'Purchase Count': playergroupscount,
                                    'Average Purchase Price': playergroupsavg,
                                    'Total Purchase Value': playergroupssum
                                   })
playergroupssummary.sort_values('Total Purchase Value', ascending=False).head(5)
```

* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting

```python
#group by items
itemgroup = data.groupby('Item Name')
itemcount = itemgroup['Purchase ID'].count()
itemprice = itemgroup['Price'].mean()
itemtotal = itemgroup['Price'].sum()
itemsummary = pd.DataFrame({'Purchase Count': itemcount,
                           'Item Price': itemprice,
                           'Total Purchase Value': itemtotal
                           })
itemsummary.sort_values('Purchase Count',ascending=False).head(5)
```
